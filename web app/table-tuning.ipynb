## Testing Suggestion Algorithm

```python
from datetime import datetime
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
filtered_df = df[df['Date'] > datetime(2023, 11, 22)]  # clearing outliers before normalizing
```

```python
u_bound = filtered_df["sensationalism_score"].quantile(0.85)    # clearing outliers again
filtered_df_2 = filtered_df[filtered_df["sensationalism_score"]<u_bound]
filtered_df_2    # keeping index to merge again in the end
```

```python
def flip_column(col):
    return -col

def rank_by_target(column, target):
    differences = abs(column - target)
    ranked_values = differences.rank(method='min', ascending=True)
    return ranked_values
```

```python
import pandas as pd
# creating dataframe for features to use for measuring
df_test = filtered_df_2[["Date", "sentiment_score", "sensationalism_score", "Readability", "bias_score", "isClickbait"]]
df_test["Date"] = pd.to_datetime(df_test["Date"])
df_test["Date"] = (df_test['Date'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')  # dates to numeric

cat_dummies = pd.get_dummies(filtered_df_2["cat_label"])   # dummies for categories
df_new = pd.concat([df_test, cat_dummies], axis=1)         # concat measure with category dummies


df_new["sensationalism_score"] = flip_column(df_new["sensationalism_score"])   # reward values that are closer to 0
df_new["bias_score"] = flip_column(df_new["bias_score"])                       # reward values that are closer to 0
df_new["isClickbait"] = flip_column(df_new["isClickbait"])                      # reward values that are closer to 0
df_new["sentiment_score"] = rank_by_target(df_new["sentiment_score"],0)         # reward values that are closer to 0
df_new["Readability"] = rank_by_target(df_new["Readability"],10)                # reward values that are closer to 10 (chosen optimal readability)
```

```python
df_normalized = (df_new - df_new.min()) / (df_new.max() - df_new.min())   # normalizing DataFrame
df_normalized.describe()                                                 # checking distributions
```

```python
# Define weights for each metric
weights = {
    "Date": 0.25,
    "Readability": 0.15,
    "sentiment_score": 0.1,
    "bias_score": 0.1,
    "isClickbait": 0.1,
    "sensationalism_score": 0.1,
    "Business and Economy": 0.03,
    "Climate and Environment": 0.02,
    "Health | Science | Technology": 0.02,
    "International News" : 0.03,
    "Law and Justice": 0.02,
    "Politics": 0.05,
    "Sports":0.03
}

df["combined_score"] = (df_normalized * pd.Series(weights)).sum(axis=1)   # final measurement

# Rank articles based on the combined score
df_ranked = df.sort_values(by="combined_score", ascending=False)
```

```python
df_ranked.to_csv("./datasets/ranked_df.csv", index=False)
```

```python
df.drop("combined_score", axis=1, inplace=True)
```

## Profile tables

```python
basis_df = df[["Date", "Text lenght", "sentiment_score", "sensationalism_score", "Readability","bias_score", "isClickbait"]]
basis_df["Date"] = pd.to_datetime(basis_df["Date"])
basis_df = basis_df[basis_df['Date'] > datetime(2023, 11, 24)]  
basis_df["Date"] = (basis_df['Date'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')  # dates to numeric
basis_df = basis_df[basis_df["Text lenght"]<100000]
basis_df
```

```python
# weights in depth explorer
weights_ide = {
    "Date": 0.2,
    "Text lenght":0.3,
    "Readability": 0.15,
    "sentiment_score": 0.05,
    "bias_score": 0.1,
    "isClickbait": 0.1,
    "sensationalism_score": 0.1,
}
```

```python
# In-Depth Explorer
filter_ide = basis_df.copy()

# text length
filter_ide = filter_ide[filter_ide["Text lenght"]>1500]  # long

# sentiment score
T_sentiment = 0.3   # neutral to positive
filter_ide["sentiment_score"] =  rank_by_target(filter_ide["sentiment_score"],T_sentiment)

#readability
T_readability = basis_df["Readability"].quantile(0.75)  # moderate to challenging
filter_ide["Readability"] =  rank_by_target(filter_ide["Readability"],T_readability)
filter_ide["Readability"] = flip_column(filter_ide["Readability"]) 

#sensationalism score
T_sensationalism = basis_df["sensationalism_score"].quantile(0.1) #low
filter_ide["sensationalism_score"] =  rank_by_target(filter_ide["sensationalism_score"],T_sensationalism)

# bias score
T_bias = basis_df["bias_score"].quantile(0.25)  # balanced
filter_ide["bias_score"] =  rank_by_target(filter_ide["bias_score"],T_bias)

# clickbait check 
filter_ide["isClickbait"] = flip_column(filter_ide["isClickbait"])   # no clickbait

ide_normalized = (filter_ide - filter_ide.min()) / (filter_ide.max() - filter_ide.min())   # normalizing DataFrame
df["combined_score"] = (ide_normalized * pd.Series(weights_ide)).sum(axis=1)   # final measurement
ide_ranked = df.sort_values(by="combined_score", ascending=False)
df.drop("combined_score", axis=1, inplace=True)
ide_ranked.head()
```

```python
ide_ranked.to_csv("./streamlit/data/In-DepthExplorer.csv", index=False)
```
