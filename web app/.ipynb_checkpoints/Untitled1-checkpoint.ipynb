```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_csv("studiengaenge.csv")
df.drop(columns=["Unnamed: 0"], inplace=True)
df
```

```python
category_keywords = {
    'Bachelor': ['B.A.', 'Bakkalaureat', 'B.A.', 'B.A.', 'Bachelor'],
    'Master': ['M.A.', 'MBA', 'M.Ed.', 'M.Eng.', 'LL.M.', "Master"],
    'Diplom': ['Diplom', 'Diplom (FH)'],
    'Staatsexamen': ['Staatsexamen', '1. Staatsprüfung', 'Staatsprüfung'],
    'Andere': ['Akademiebrief', 'Lehramt', 'Meisterklasse', 'Magister', '1. Theologisches Examen', 'Kirchliches Examen']
}

degree_categories = []

for degree in df["Abschluss"]:
    assigned_category = 'Andere' 

    for category, keywords in category_keywords.items():
        if any(keyword in degree for keyword in keywords):
            assigned_category = category
            break
    degree_categories.append(assigned_category)

df["Abschluss_label"] = degree_categories
df
```
