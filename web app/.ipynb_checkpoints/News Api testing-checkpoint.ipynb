```python
import pandas as pd
import numpy as np

import requests
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
api_endpoint = 'https://newsapi.org/v2/top-headlines'
api_key = 'fb09799753ef4d99983d6685b364335c'
def get_articles(country, category):
    params = {
        'country': country,
        'category': category,
        'pageSize': 100,
        'apiKey': api_key
    }

    r = requests.get(api_endpoint, params)
    news_data = r.json()
    articles = news_data.get('articles', [])
    df = pd.DataFrame(articles)
    df["category"] = category
    df["date"] = df["publishedAt"].apply(lambda x: x.split("T")[0])
    df["time"] = df["publishedAt"].apply(lambda x: x.split("T")[1])
    for i in range(len(df["source"])):
        df["source"][i] = df["source"][i]["name"]
    df.drop(["urlToImage","publishedAt"], axis=1, inplace=True)
    return df
```

```python
countries = ['us', 'gb', 'de']
categories = ["business", "entertainment", "general", "health", "science", "sports", "technology"]
dfs = []
for country in countries:
    for cat in categories:
        dfs.append(get_articles(country, cat))
articles = pd.concat(dfs, ignore_index=True)
len(articles)
```
